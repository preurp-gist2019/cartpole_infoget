```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import json
import os
from glob import glob
```

```python
json_dir = [y for x in os.walk('/preurp/deep-q-learning/logs/') for y in glob(os.path.join(x[0], '*.json'))]
```

```python
json_dir.sort()
a = json_dir
```

```python
with open(json_dir[0]) as json_file:
    data = json.load(json_file)
```

```python
epi_len = len(data.keys()) - 2
acc_step = np.zeros(0)
loss_step = np.zeros(0)
acc_epi = np.zeros(epi_len)
loss_epi = np.zeros(epi_len)

for i in range(epi_len):
    epi = str(i)
    data_arr = np.array(data[epi])
    
    acc_step = np.concatenate((acc_step, data_arr[:,0]), axis=0)
    loss_step = np.concatenate((loss_step, data_arr[:,1]), axis=0)
    
    acc_epi[i] = np.mean(data_arr[:,0])
    loss_epi[i] = np.mean(data_arr[:,1])
    
```

```python
fig = plt.figure(figsize=(12,12))
axes = {}
for i in range(1,5):
    axes[i] = fig.add_subplot(2,2,i)

axes[1].plot(acc_step)
axes[1].set_xlabel("Training Steps")
axes[1].set_ylabel("Training Accuracy")
axes[1].set_title("Training Accuracy (by Step)")


axes[2].plot(loss_step)
axes[2].set_xlabel("Training Steps")
axes[2].set_ylabel("Training Loss (MSE)")
axes[2].set_title("Training Loss (by Step)")


axes[3].plot(acc_epi)
axes[3].set_xlabel("Training Episodes")
axes[3].set_ylabel("Training Accuracy")
axes[3].set_title("Training Accuracy (by Episode)")

axes[4].plot(loss_epi)
axes[4].set_xlabel("Training Episodes")
axes[4].set_ylabel("Training Loss (MSE)")
axes[4].set_title("Training Loss (by Episode)")

plt.show()
```

```python
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

x_axis = np.arange(epi_len)
ax1.plot(data['score'])
ax1.set_xticks(x_axis)
ax1.set_title("Scores along Episode")
ax1.set_xlabel("Episode")
ax1.set_ylabel("Score")

ax2.plot(data['epsilon'])
ax2.set_xticks(x_axis)
ax2.set_title("Epsilons along Episode")
ax2.set_xlabel("Episode")
ax2.set_ylabel("Epsilon")

plt.show()
```
